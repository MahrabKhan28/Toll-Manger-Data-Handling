```python
import pandas as pd
import numpy as np

# 1. Setting up 1,000 mock rows
np.random.seed(42)
n_rows = 1000

# 2. Creating realistic distributions using NumPy
timestamps = pd.date_range(start='2026-06-01 00:00:00', periods=n_rows, freq='15min')
states = ['CA', 'NY', 'TX', 'FL', 'IL']
plates = [f"{np.random.choice(states)}-{np.random.randint(1000, 9999)}" for _ in range(n_rows)]
speeds = np.random.normal(loc=65, scale=12, size=n_rows).round(1) # Mean 65mph
base_tolls = np.random.choice([10.00, 12.00, 15.00], size=n_rows, p=[0.4, 0.4, 0.2])
vehicle_types = np.random.choice(['Sedan', 'SUV', 'Truck', 'EV'], size=n_rows, p=[0.4, 0.3, 0.2, 0.1])

# 3. Injecting a few missing values (NaNs) to challenge your data cleaning skills!
speeds[np.random.choice(n_rows, size=15, replace=False)] = np.nan

# 4. Assembling and saving
df = pd.DataFrame({
    'Timestamp': timestamps, 'Vehicle_Plate': plates, 
    'Recorded_Speed': speeds, 'Base_Toll': base_tolls, 'Vehicle_Type': vehicle_types
})
df.to_csv('traffic_logs.csv', index=False)
print("Success! 'traffic_logs.csv' with 1,000 rows has been created in your folder.")
```
